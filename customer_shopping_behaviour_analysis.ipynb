#### Now we have to fill the missing rating with some value with median of that category
#### why median and not mean?

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

#### Now we don't have any null value
But we need to change the names of columns in lowercase and also replace space with underscore

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

All the columns except for purchase_amount_(usd) are proper. We need to remove these (,)

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount_usd'})
```

```python
df.columns
```

Now create a new column age_group

```python
labels = ['Young Adult', 'Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels = labels)
```

```python
df[['age','age_group']].head(10)
```

Now we have to create a new column 
#### purchase_frequency_days Why? - to work with numerical data

```python
frequency_mapping = { 
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annualy':365,
    'Every 3 Months':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

Now the columns promo_code_used and discount_applied have same values.

we must delete one column from df

```python
df = df.drop('promo_code_used',axis=1)
```
